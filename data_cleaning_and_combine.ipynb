```python
import pandas as pd
import difflib
```

# CHE

```python
che = pd.read_csv('0807-che-cleaning.csv')
che.head(3)
```

```python
che['Name'] = che['Name'].str.replace('U.', 'University')
che['Name'] = che['Name'].str.replace(',', '-')
che['Name'] = che['Name'].str.replace('at', '-')
che['Name'] = che['Name'].str.replace('.', '')
che['Name'] = che['Name'].str.replace('City University of New York', 'CUNY')
che['Name'] = che['Name'].str.replace('St-e', 'State')
che['Name'] = che['Name'].str.replace('Dram-ic', 'Dramatic')
che['Name'] = che['Name'].str.replace('Cincinn-i', 'Cincinnati')
che['Name'] = che['Name'].str.replace('B-es', 'Bates')
che['Name'] = che['Name'].str.replace('P-h', 'Path')
```

```python
def remove_pattern(x):
    start = x.find('(')
    if start != -1:
        result = x[:start]
    else:
        result = x
    return result
```

```python
che['Name'] = che['Name'].apply(remove_pattern)
```

# University Information

```python
university = pd.read_csv('overall_0818.csv')
university['county'] = university['county'].str.replace('County', '')
university['county'] = university['county'].str.strip()
university.head(3)
```

```python
university.info()
```

```python
#define function difference 
def best_guess ( usn_name ):
  """match possible univeristies in the two dataset 

  Args: 
    usn_name : str 
    the school name you trying to guess 

  Result : str
    Most close university name 

  """
  if usn_name in list( university['CHE_name'] ):
      return usn_name
  else:
      close_ones = difflib.get_close_matches( usn_name, list( university['CHE_name'] ), 1)
      if len( close_ones ) > 0:
          return close_ones[0]
      else:
          return np.nan
```

```python
che['institution'] = che['Name'].apply( best_guess )
che
```

```python
university_che = pd.merge(che,
                          university,
                          left_on=['institution', 'State'],
                          right_on=['CHE_name', 'state'])
university_che.drop(columns = ['Name','CHE_name','CNES_name','state','control'],inplace = True)
university_che.info()
```

```python
covid['abbrev'] = covid['state'].map(us_state_abbrev)
```

```python
covid = covid[(covid['date'] == '2020-03-30') | (covid['date'] == '2020-04-30')
              | (covid['date'] == '2020-05-30') |
              (covid['date'] == '2020-06-30') |
              (covid['date'] == '2020-07-30')]
covid['month'] = covid['date'].dt.month
covid
```

```python
covid = covid.pivot_table(index=['abbrev', 'county'],
                          columns=['month'],
                          values=['cases', 'deaths'],
                          aggfunc='sum',
                          fill_value=0).reset_index()
covid
```

```python
covid.columns = [
    'state', 'county', 'cases_mar', 'cases_apr', 'cases_may', 'cases_june',
    'cases_july', 'death_mar', 'death_apr', 'death_may', 'death_june',
    'death_july'
]
covid
```

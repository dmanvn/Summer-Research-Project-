```
#import packages 
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 


pd.options.display.max_rows = 999
plt.style.use('fivethirtyeight')
%matplotlib inline 
```

```
# import dataset 
df = pd.read_csv('/content/drive/My Drive/Summer-Research/data/Covid_19_By_states.csv',usecols = ['date','state','positive','hospitalizedCumulative','inIcuCumulative','onVentilatorCumulative','recovered','death'],parse_dates=['date'])
df['month'] = df['date'].dt.month
df.sort_values('month',ascending=True,inplace = True)
df.head()
```

```
# only choose march, april, may and june
df = df[(df.month ==3)|(df.month==4)|(df.month==5)|(df.month==6)]
```

```
# pivot table to aggregate data 
agg_df = df.pivot_table(index = 'state',columns = 'month',aggfunc = 'sum',fill_value=0,)
agg_df.drop(index= 'AS',inplace = True)
agg_df
```

```
df2 = pd.read_csv('/content/drive/My Drive/Summer-Research/data/cleaned_chronicle.csv',index_col=[0])
df2.head()
```

```
df_new = pd.merge(df2,agg_df,left_on='State',right_index=True,how = 'right').reset_index()
df_new.drop(columns = 'index',inplace = True)
df_new.head()
```

```
#change columns name to be more clear 
columns_name = ['Institution','Control','State','Category','Plan','Lat','Long',
                           'March_Death','April_Death','May_Death','June_Death',
                           'March_Hospitalized','April_Hospitalized','May_Hospitalized','June_Hospitalized',
                           'March_InICU','April_InICU','May_InICU','June_InICU',
                           'March_onVentilatorCumulative','April_onVentilatorCumulative','May_onVentilatorCumulative','June_onVentilatorCumulative',
                           'March_Positive','April_Positive','May_Positive','June_Positive',
                           'March_Recovered','April_Recovered','May_Recovered','June_Recovered']
```

```
# new data frame 
df_new.columns = columns_name
df_new.head()
```

```
df_new.to_csv('combine_covid_19.csv',index = False)
```

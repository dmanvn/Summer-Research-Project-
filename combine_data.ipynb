```
import pandas as pd
```

```
state = pd.read_csv('/content/state.csv')
state
```

```
election = pd.read_csv('/content/drive/My Drive/SUMMER RESEARCH GROUP/final_data /election_2016_counties.csv')
election
```

```
politic = pd.merge(state,election,left_on = 'State',right_on = 'state')
```

```
politic.drop(columns = ['State','state'],inplace = True)
politic.rename(columns={'Name':'state'},inplace = True)
politic = politic.append({'state':'AK','trump_percent':0.5281,'clinton_percent':0.3764},ignore_index=True)
politic
```

```
econ = pd.read_csv('/content/drive/My Drive/SUMMER RESEARCH GROUP/final_data /econ.csv')
econ
```

```
state = []
for i in range(len(econ)):
  state.append(econ['state'][i].strip())
```

```
econ['state'] = state
```

```
econ_politic = pd.merge(econ,politic,how = 'left')
econ_politic
```

```
covid = pd.read_csv('/content/drive/My Drive/SUMMER RESEARCH GROUP/final_data /covid-month-counties.csv')
covid
```

```
econ_politic_covid = pd.merge(econ_politic,covid)
econ_politic_covid
```

```
university = pd.read_csv('/content/drive/My Drive/SUMMER RESEARCH GROUP/final_data /overall_0818.csv')
university['county'] = university['county'].str.replace('County','')
university['county'] = university['county'].str.strip()
university
```

```
university_covid_econ_politic = pd.merge(university,econ_politic_covid,how = 'left',left_on = ['state','county'],right_on = ['state','county'])
university_covid_econ_politic.info()
```

```
che = pd.read_csv('/content/drive/My Drive/SUMMER RESEARCH GROUP/final_data /0807-che-cleaning.csv')
che
```

```
final_data = pd.merge(che,university_covid_econ_politic,left_on = ['Name','State'],right_on = ['CHE_name','state'])
final_data
```

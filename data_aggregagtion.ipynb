```
#import packages 
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 


pd.options.display.max_rows = 999
plt.style.use('fivethirtyeight')
%matplotlib inline 
```

```
# import covid-19 data 
covid = pd.read_csv('/content/drive/My Drive/Summer-Research/data/Covid_19_By_states.csv',usecols = ['date','state','positive','hospitalizedCumulative','inIcuCumulative','onVentilatorCumulative','recovered','death'],parse_dates=['date'])
covid['month'] = covid['date'].dt.month
covid.sort_values('month',ascending=True,inplace = True)
covid.head()
```

```
# only choose march, april, may and june
covid = covid[(covid.month ==3)|(covid.month==4)|(covid.month==5)|(covid.month==6)]
```

```
# pivot table to aggregate data 
agg_covid = covid.pivot_table(index = 'state',columns = 'month',aggfunc = 'sum',fill_value=0)
agg_covid.drop(index= 'AS',inplace = True)
agg_covid
```

```
# combine cleaned CHE data 
df2 = pd.read_csv('/content/drive/My Drive/Summer-Research/data/cleaned_chronicle.csv',index_col=[0])
df2.head()
```

```
df_new = pd.merge(df2,agg_covid,left_on='State',right_index=True,how = 'right').reset_index()
df_new.drop(columns = 'index',inplace = True)
df_new.head()
```

```
#change columns name to be more clear 
columns_name = ['Institution','Control','State','Category','Plan','Lat','Long',
                           'March_Death','April_Death','May_Death','June_Death',
                           'March_Hospitalized','April_Hospitalized','May_Hospitalized','June_Hospitalized',
                           'March_InICU','April_InICU','May_InICU','June_InICU',
                           'March_onVentilatorCumulative','April_onVentilatorCumulative','May_onVentilatorCumulative','June_onVentilatorCumulative',
                           'March_Positive','April_Positive','May_Positive','June_Positive',
                           'March_Recovered','April_Recovered','May_Recovered','June_Recovered']
```

```
# new data frame 
df_new.columns = columns_name
df_new.head()
```

```
# drop some columns with missing 
df_new.dropna(how = 'any',inplace = True)
```

```
# import politic data 
politic = pd.read_csv('/content/drive/My Drive/Summer-Research/data/president.csv',usecols=['year','state_po','party'])
politic = politic[politic.year>=2000]
politic.head()
```

```
#limit party to democrat,republican,libertarian,green,constitution party 
politic = politic[(politic.party =='republican')|(politic.party == 'democrat')|(politic.party == 'libertarian')|(politic.party == 'green') |(politic.party == 'constitution party')]
politic = politic.pivot_table(index = 'state_po',columns = 'party',aggfunc='count',fill_value=0)
politic.columns = ['constitution_party_numbers','democrat_numbers','green_numbers','libertarian_numbers','republican_numbers']
politic
```

```
# combine politic data with clean 
df_new = pd.merge(df_new,politic,how = 'left',left_on = 'State',right_index = True)
df_new.head()
```

```
# import economic data 
gdp = pd.read_csv('/content/drive/My Drive/Summer-Research/data/gdp.csv',
                  skiprows=4)
gdp.drop(columns = 'GeoFips',inplace = True)
gdp.drop(index = 0,inplace = True)
gdp.columns = ['State','2015_gdp','2016_gdp','2017_gdp','2018_gdp','2019_gdp']
gdp.head()
```

```
# import income 
income = pd.read_csv('/content/drive/My Drive/Summer-Research/data/income.csv',skiprows=4)
income.drop(columns = 'GeoFips',inplace = True)
income.drop(index = 0,inplace = True)
income.columns = ['State','2015_personal_income','2016_personal_income','2017_personal_income','2018_personal_income','2019_personal_income']

# change data with * 
income['State'].iloc[11] = 'Hawaii'
income['State'].iloc[1] = 'Alaska'

# income 
income.head()
```

```
# import employment data 
emp = pd.read_csv('/content/drive/My Drive/Summer-Research/data/total-employement.csv',skiprows=4)
emp.drop(columns = 'GeoFips',inplace = True)
emp.drop(index = 0,inplace = True)
emp.columns = ['State','2015_emp','2016_emp','2017_emp','2018_emp','2019_emp']

# clean data with other value 
emp.replace('(NA)',np.NaN,inplace = True)
emp.drop(columns = '2019_emp',inplace = True)
emp['State'].iloc[1] ='Alaska'
emp['State'].iloc[11]='Hawaii'

# employment data 
emp.head()
```

```
# combine data 
eco = pd.merge(gdp,income)
eco_big = pd.merge(eco,emp)
eco_big.dropna(how = 'all',inplace = True)
eco_big.head()
```

```
# import state data to change data name to alias 
state = pd.read_csv('/content/drive/My Drive/第三个学期/MA 705/Project_1/dataset/statename.csv')
state['State'].iloc[17]='Louisiana'
state.head()
```

```
# merge two data 
econ_new = pd.merge(state,eco_big)
econ_new.drop(columns = 'State',inplace = True)
econ_new.rename(columns={'Name':'State'},inplace = True)
econ_new.head()
```

```
# merge econ_new with previous merge data 
data = pd.merge(df_new,econ_new,how='left')
data.head()
```
